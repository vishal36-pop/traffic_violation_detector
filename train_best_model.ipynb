# Train Best YOLOv8 Model
This notebook trains the YOLOv8l (Large) model for exactly 40 epochs.
It uses a batch size of 8 so it fits in the 6 GB VRAM on the RTX 4050. The resulting weights size will be ~83 MB, safely under the 250 MB course constraint.

```python
import os
import torch
from ultralytics import YOLO

# Check available GPU
device = "0" if torch.cuda.is_available() else "cpu"
if device == "0":
    print(f"Using GPU: {torch.cuda.get_device_name(0)}")
else:
    print("Using CPU")
```

```python
# Initialize the best fitting model (YOLOv8 Large)
model = YOLO("yolov8l.pt")
```

```python
# Train the model for 40 epochs
results = model.train(
    data="C:/Users/VishalReddyK/OneDrive/Documents/semester 4/Computer Vision/project/traffic_violation_detector/data/combined_dataset.yaml",
    epochs=40,
    batch=8,           # Lowered to 8 to avoid Out-of-Memory (OOM) with the YOLOv8l model
    imgsz=640,
    device=device,
    project="models",
    name="combined_detector_best",
    exist_ok=True,
    optimizer="AdamW",
    lr0=0.001,
    # Small object augmentations (for helmets and plates)
    mosaic=1.0,
    copy_paste=0.15,
    mixup=0.05
)
```

```python
# Run validation to get final metrics
metrics = model.val()
print("\nTraining complete! See charts in the models/combined_detector_best folder.")
```
